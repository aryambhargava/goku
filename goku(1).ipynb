```python
blur=cv2.GaussianBlur(inv,ksize=(21,21),sigmaX=0,sigmaY=0)
```

```python
import cv2

# Load your image and mask here
image = cv2.imread("'goku.jpg'")
mask = cv2.imread("'goku.jpg'", cv2.IMREAD_GRAYSCALE)


```

```python
# Define the dodge lambda function
dodge = lambda image, mask: cv2.divide(image, 255 - mask, scale=256)

```
